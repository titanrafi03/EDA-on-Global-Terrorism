# Make Visuals with Gathered Data: Data Visualization

## Number of Terrorist Activities Each Year:

```python
x_year = gt['Year'].unique()
y_count_years = gt['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'rocket')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

## Terrorist Activities by Region in each Year through Area Plot:

```python
pd.crosstab(gt.Year, gt.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
gt['Wounded'] = gt['Wounded'].fillna(0).astype(int)
gt['Killed'] = gt['Killed'].fillna(0).astype(int)
gt['casualities'] = gt['Killed'] + gt['Wounded']
```

### Values are Sorted by the Top 40 Worst Terror Attacks as to Keep the Heatmap Simple and Easy to Visualize:

```python
gt1 = gt.sort_values(by='casualities',ascending=False)[:40]
```

```python
heat=gt1.pivot_table(index='Country',columns='Year',values='casualities')
heat.fillna(0,inplace=True)
```

```python
heat.head()
```

```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
colorscale = [[0, '#edf8fb'], [.3, '#00BFFF'],  [.6, '#8856a7'],  [1, '#810f7c']]
heatmap = go.Heatmap(z=heat.values, x=heat.columns, y=heat.index, colorscale=colorscale)
data = [heatmap]
layout = go.Layout(
    title='Top 40 Worst Terror Attacks in History from 1982 to 2017',
    xaxis = dict(ticks='', nticks=20),
    yaxis = dict(ticks='')
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='heatmap',show_link=False)
```

# Total Number of People Killed in Terror Attack:

```python
killData = gt.loc[:,'Killed']
print('Number of people killed by terror attack:', int(sum(killData.dropna())))
```

# Let's Analyze the types of Attacks Involved in these Deaths:

```python
attackData = gt.loc[:,'AttackType']
typeKillData = pd.concat([attackData, killData], axis=1)
```

```python
typeKillData.head()
```

```python
typeKillFormatData = typeKillData.pivot_table(columns='AttackType', values='Killed', aggfunc='sum')
typeKillFormatData
```

```python
typeKillFormatData.info()
```

```python
import matplotlib.pyplot as plt

labels = typeKillFormatData.columns.tolist()
transpoze = typeKillFormatData.T
values = [x[0] for x in transpoze.values.tolist()]

fig, ax = plt.subplots(figsize=(20, 20), subplot_kw=dict(aspect="equal"))
plt.pie(values, startangle=90, autopct='%.2f%%')
plt.title('Types of terrorist attacks that cause deaths')
plt.legend(labels, loc='upper right', bbox_to_anchor=(1.3, 0.9), fontsize=15)
plt.show()
```
